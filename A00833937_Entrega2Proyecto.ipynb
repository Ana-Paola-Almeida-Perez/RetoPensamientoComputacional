```python
empleados = []

def alta_empleado():

  global empleados

  id = int(input("Teclea el id del empleado: "))
  nombre = str(input("Teclea el nombre del empleado: "))
  edad = int(input("Teclea edad del empleado: "))
  peso = int(input("Teclea el peso (Kg): "))
  estatura = float(input("Teclea la estatura (Mts.): "))
  diabetes = int(input("Diabetes 1-Sí 0-No ? : "))
  hipertension = int(input("Hipertensión 1-Sí 0-No ? : "))
  corazon = int(input("Enfermedad corazón 1-Sí 0-No ? : "))
  cancer = int(input("Cáncer 1-Sí 0-No ? : "))
  tabaquismo = int(input("Tabaquismo 1-Sí 0-No ? : "))
  vacuna = int(input("Vacuna Covid-19 1-Sí 0-No ? : "))
  departamento = int(input("Teclea el número de departamento: "))

  empleado =  [id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento]
  print(f"Empleado dado de alta {empleado}")
  empleados.append(empleado)




def calcular_imc():

  global empleados

  string = "Clasificación:\t                  IMC:\t                  ID:\t         Nombre:\t         Enfermedades:\t                                                                                                                                                                     Vacuna"
  print(f'{string}')

  delgadez_severa = 0
  delgadez_moderada = 0
  delgadez_aceptable = 0
  peso_normal = 0
  pre_obeso = 0
  obeso_tipo1 = 0
  obeso_tipo2 = 0
  obeso_tipo3 = 0

  for empleado in empleados:
    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado
    imc = round((peso) / ((estatura)**2),2)



    if imc < 18.50:
      if imc < 16.00:
        clasificacion = "Delgadez Severa"
        delgadez_severa = delgadez_severa + 1
        
      elif imc >= 16.00 and imc < 17.00:
        clasificacion = "Delgadez moderada"
        delgadez_moderada = delgadez_moderada + 1
       
      elif imc >= 17.00 and imc < 18.50:
        clasificacion = "Delgadez aceptable"
        delgadez_aceptable = delgadez_aceptable + 1
        

    elif imc >= 18.50 and imc < 25.00:
      clasificacion = "Peso normal"
      peso_normal = peso_normal + 1
    

    elif imc >= 25.00 and imc < 30.00:
      clasificacion = "pre-obeso (riesgo)"
      pre_obeso = pre_obeso + 1
      
      
    elif imc >= 30.00:
      if imc >= 30.00 and imc < 35.00:
        clasificacion = "Obeso tipo 1 (Riesgo moderado)"
        obeso_tipo1 = obeso_tipo1 + 1
    
      elif imc >= 35.00 and imc < 40.00:
        clasificacion = "Obeso tipo 2 (Riesgo Severo)"
        obeso_tipo2 = obeso_tipo2 + 1

      elif imc >= 40.00:
        clasificacion = "Obeso tipo 3 (Riesgo muy severo)"
        obeso_tipo3 = obeso_tipo3 + 1     

    print(f'{clasificacion:<35}{imc:<25}{id:<8}{nombre:<30}{diabetes:<10}{hipertension:<10}{corazon:<10}{cancer:<10}{tabaquismo:<10}{vacuna:<15}')

  porcentaje_delgadez_severa = (delgadez_severa)*100 / len(empleados) 
  porcentaje_delgadez_moderada = (delgadez_moderada)*100 / len(empleados)
  porcentaje_delgadez_aceptable = (delgadez_aceptable)*100 / len(empleados)
  porcentaje_peso_normal = (peso_normal)*100 / len(empleados)
  porcentaje_pre_obeso = (pre_obeso)*100 / len(empleados)
  porcentaje_obeso_tipo1 = (obeso_tipo1)*100 / len(empleados)
  porcentaje_obeso_tipo2 = (obeso_tipo2)*100 / len(empleados)
  porcentaje_obeso_tipo3 = (obeso_tipo3)*100 / len(empleados)
  string_2 = "\t                                                                        Diagnóstico:\t                  Cantidad:\t                  Porcentaje:"
  print(f'{string_2}')
  print(f'\t                                                                       Delgadez severa\t                        {delgadez_severa:<30}{porcentaje_delgadez_severa}%')
  print(f'\t                                                                       Delgadez moderada\t                {delgadez_moderada:<30}{porcentaje_delgadez_moderada}%')
  print(f'\t                                                                       Delgadez aceptable\t                {delgadez_aceptable:<30}{porcentaje_delgadez_aceptable}%')
  print(f'\t                                                                       Peso normal\t                        {peso_normal:<30}{porcentaje_peso_normal}%')
  print(f'\t                                                                       Pre-obeso\t                        {pre_obeso:<30}{porcentaje_pre_obeso}%')
  print(f'\t                                                                       Obeso tipo 1\t                        {obeso_tipo1:<30}{porcentaje_obeso_tipo1}%')
  print(f'\t                                                                       Obeso tipo 2\t                        {obeso_tipo2:<30}{porcentaje_obeso_tipo2}%')
  print(f'\t                                                                       Obeso tipo 3\t                        {obeso_tipo3:<30}{porcentaje_obeso_tipo3}%')

  #print(f'{clasificacion:<35}{imc:<25}{id:<8}{nombre:<30}{diabetes:<10}{hipertension:<10}{corazon:<10}{cancer:<10}{tabaquismo:<10}{vacuna:<15}')        
 
    

  
def actualizar_empleado():

  global empleados 
  
  print("Bienvenido a la opción para actualizar empleado, por favor sigue las instrucciones \nA continuación se despliega la lista de preguntas guardadas\n")

  for empleado in empleados:

    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    print(f"ID del empleado: {empleado[0]}  Nombre del empleado: {empleado[1]}  Edad del empleado: {empleado[2]}  Peso del empleado: {empleado[3]}  Estatura del empleado: {empleado[4]}  Diabetes (1 = Sí, 0= No): {empleado[5]} Hipertensión (1= Sí, 0= No): {empleado[6]}  Problemas del corazón (1=Sí, 0= No): {empleado[7]}  Cancer (1=Sí, 0= No): {empleado[8]} Tabaquismo (1=Sí, 0= No): {empleado[9]} Vacuna (1=Sí, 0= No): {empleado[10]}  Departamento: {empleado[11]}\n")
  
  id_cambio = str(input("Ingresa el id del empleado cuya información quieras modificar: "))

  for empleado in empleados:
    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    if id_cambio == id:
      existe = "si"

    else:
      existe = "no" 
  
  if existe == "no":
    print("El empleado no existe")

  else:
    print(f'Empleado encontrado, ID:    {empleado[0]}')
    for x in range(len(empleado)):
      id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado
      if empleado[x] == id:
        empleado[x] = int(input("Teclea el id del empleado: "))
        empleado.append(int(empleado[x]))

      if empleado[x] == nombre:
        empleado[x] = str(input("Teclea el nombre del empleado: "))
      if empleado[x] == edad:
        empleado[x] = int(input("Teclea edad del empleado: "))
      if empleado[x] == peso:
        empleado[x] = int(input("Teclea el peso (Kg): "))
      if empleado[x] == estatura:
        empleado[x] = float(input("Teclea la estatura (Mts.): "))
      if empleado[x] == diabetes:
        empleado[x] = int(input("Diabetes 1-Sí 0-No ? : "))
      if empleado[x] == hipertension:
        empleado[x] = int(input("Hipertensión 1-Sí 0-No ? : "))
      if empleado[x] == corazon:
        empleado[x] = int(input("Enfermedad corazón 1-Sí 0-No ? : "))
      if empleado[x] == cancer:
        empleado[x] = int(input("Cáncer 1-Sí 0-No ? : "))
      if empleado[x] == tabaquismo:
        empleado[x] = int(input("Tabaquismo 1-Sí 0-No ? : ")
      if empleado[x] == vacuna: 
        empleado[x] = int(input("Vacuna Covid-19 1-Sí 0-No ? : "))
      if empleado[x] == departamento:
        empleado[x] = int(input("Teclea el número de departamento: ")) 
    print(empleado)   
    

def consultar_empleado():
  global empleados 
  
  print("Bienvenido a la opción para consultar un empleado, por favor sigue las instrucciones \nA continuación se despliega la lista de preguntas guardadas\n")

  for empleado in empleados:

    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    print(f"ID del empleado: {empleado[0]}  Nombre del empleado: {empleado[1]}  Edad del empleado: {empleado[2]}  Peso del empleado: {empleado[3]}  Estatura del empleado: {empleado[4]}  Diabetes (1 = Sí, 0= No): {empleado[5]} Hipertensión (1= Sí, 0= No): {empleado[6]}  Problemas del corazón (1=Sí, 0= No): {empleado[7]}  Cancer (1=Sí, 0= No): {empleado[8]} Tabaquismo (1=Sí, 0= No): {empleado[9]} Vacuna (1=Sí, 0= No): {empleado[10]}  Departamento: {empleado[11]}\n")
  
  id_cambio = str(input("Ingresa el id del empleado cuya información quieras modificar: "))

  for empleado in empleados:
    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    if id_cambio == id:
      existe = "si"

    else:
      existe = "no" 
  
  if existe == "no":
    print("El empleado no existe")

  else:
    print(f'Empleado encontrado, {empleado}')

def crear_reporte_empleados():
  global empleados



def salir():
  print("Gracias por usar el sistema")
  si = False

def menu():

  print("Sistema de Salud y Bienestar del Empleado")
  print("1. Alta de empleado")
  print("2. Calcular imc de empleados")
  print("3. Cambia información del empleaddo")
  print("4. Consulta un empleado")
  print("5. Reporte de todos los empleados")
  print("6. Salir")
  opcion = int(input("Teclea la opcion: "))
  return opcion

def main():

  si = True 
  #matriz = []
  
  while si:

    opcion = menu()

    if opcion < 1 and opcion > 7:
      print("Opción incorrecta")
      menu()

    if opcion == 1:
      alta_empleado()

    elif opcion == 2:
      calcular_imc()
   
    elif opcion == 3: 
      actualizar_empleado()

    elif opcion == 4:
      consultar_empleado()
      
    elif opcion == 5:
      crear_reporte_empleados()

    elif opcion == 6:
      si = salir()
  
       
main()
```

```python
empleados = []

def alta_empleado():

  global empleados

  id = int(input("Teclea el id del empleado: "))
  nombre = str(input("Teclea el nombre del empleado: "))
  edad = int(input("Teclea edad del empleado: "))
  peso = int(input("Teclea el peso (Kg): "))
  estatura = float(input("Teclea la estatura (Mts.): "))
  diabetes = int(input("Diabetes 1-Sí 0-No ? : "))
  hipertension = int(input("Hipertensión 1-Sí 0-No ? : "))
  corazon = int(input("Enfermedad corazón 1-Sí 0-No ? : "))
  cancer = int(input("Cáncer 1-Sí 0-No ? : "))
  tabaquismo = int(input("Tabaquismo 1-Sí 0-No ? : "))
  vacuna = int(input("Vacuna Covid-19 1-Sí 0-No ? : "))
  departamento = int(input("Teclea el número de departamento: "))

  empleado =  [id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento]
  print(f"Empleado dado de alta {empleado}")
  empleados.append(empleado)




def calcular_imc():

  global empleados

  string = "Clasificación:\t                  IMC:\t                  ID:\t         Nombre:\t         Enfermedades:\t                                                                                                                                                                     Vacuna"
  print(f'{string}')

  delgadez_severa = 0
  delgadez_moderada = 0
  delgadez_aceptable = 0
  peso_normal = 0
  pre_obeso = 0
  obeso_tipo1 = 0
  obeso_tipo2 = 0
  obeso_tipo3 = 0

  for empleado in empleados:
    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado
    imc = round((peso) / ((estatura)**2),2)



    if imc < 18.50:
      if imc < 16.00:
        clasificacion = "Delgadez Severa"
        delgadez_severa = delgadez_severa + 1
        
      elif imc >= 16.00 and imc < 17.00:
        clasificacion = "Delgadez moderada"
        delgadez_moderada = delgadez_moderada + 1
       
      elif imc >= 17.00 and imc < 18.50:
        clasificacion = "Delgadez aceptable"
        delgadez_aceptable = delgadez_aceptable + 1
        

    elif imc >= 18.50 and imc < 25.00:
      clasificacion = "Peso normal"
      peso_normal = peso_normal + 1
    

    elif imc >= 25.00 and imc < 30.00:
      clasificacion = "pre-obeso (riesgo)"
      pre_obeso = pre_obeso + 1
      
      
    elif imc >= 30.00:
      if imc >= 30.00 and imc < 35.00:
        clasificacion = "Obeso tipo 1 (Riesgo moderado)"
        obeso_tipo1 = obeso_tipo1 + 1
    
      elif imc >= 35.00 and imc < 40.00:
        clasificacion = "Obeso tipo 2 (Riesgo Severo)"
        obeso_tipo2 = obeso_tipo2 + 1

      elif imc >= 40.00:
        clasificacion = "Obeso tipo 3 (Riesgo muy severo)"
        obeso_tipo3 = obeso_tipo3 + 1     

    print(f'{clasificacion:<35}{imc:<25}{id:<8}{nombre:<30}{diabetes:<10}{hipertension:<10}{corazon:<10}{cancer:<10}{tabaquismo:<10}{vacuna:<15}')

  porcentaje_delgadez_severa = (delgadez_severa)*100 / len(empleados) 
  porcentaje_delgadez_moderada = (delgadez_moderada)*100 / len(empleados)
  porcentaje_delgadez_aceptable = (delgadez_aceptable)*100 / len(empleados)
  porcentaje_peso_normal = (peso_normal)*100 / len(empleados)
  porcentaje_pre_obeso = (pre_obeso)*100 / len(empleados)
  porcentaje_obeso_tipo1 = (obeso_tipo1)*100 / len(empleados)
  porcentaje_obeso_tipo2 = (obeso_tipo2)*100 / len(empleados)
  porcentaje_obeso_tipo3 = (obeso_tipo3)*100 / len(empleados)
  string_2 = "\t                                                                        Diagnóstico:\t                  Cantidad:\t                  Porcentaje:"
  print(f'{string_2}')
  print(f'\t                                                                       Delgadez severa\t                        {delgadez_severa:<30}{porcentaje_delgadez_severa}%')
  print(f'\t                                                                       Delgadez moderada\t                {delgadez_moderada:<30}{porcentaje_delgadez_moderada}%')
  print(f'\t                                                                       Delgadez aceptable\t                {delgadez_aceptable:<30}{porcentaje_delgadez_aceptable}%')
  print(f'\t                                                                       Peso normal\t                        {peso_normal:<30}{porcentaje_peso_normal}%')
  print(f'\t                                                                       Pre-obeso\t                        {pre_obeso:<30}{porcentaje_pre_obeso}%')
  print(f'\t                                                                       Obeso tipo 1\t                        {obeso_tipo1:<30}{porcentaje_obeso_tipo1}%')
  print(f'\t                                                                       Obeso tipo 2\t                        {obeso_tipo2:<30}{porcentaje_obeso_tipo2}%')
  print(f'\t                                                                       Obeso tipo 3\t                        {obeso_tipo3:<30}{porcentaje_obeso_tipo3}%')

  #print(f'{clasificacion:<35}{imc:<25}{id:<8}{nombre:<30}{diabetes:<10}{hipertension:<10}{corazon:<10}{cancer:<10}{tabaquismo:<10}{vacuna:<15}')        
 
    

  
def actualizar_empleado():

  global empleados 
  
  print("Bienvenido a la opción para actualizar empleado, por favor sigue las instrucciones \nA continuación se despliega la lista de preguntas guardadas\n")

  for empleado in empleados:

    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    print(f"ID del empleado: {empleado[0]}  Nombre del empleado: {empleado[1]}  Edad del empleado: {empleado[2]}  Peso del empleado: {empleado[3]}  Estatura del empleado: {empleado[4]}  Diabetes (1 = Sí, 0= No): {empleado[5]} Hipertensión (1= Sí, 0= No): {empleado[6]}  Problemas del corazón (1=Sí, 0= No): {empleado[7]}  Cancer (1=Sí, 0= No): {empleado[8]} Tabaquismo (1=Sí, 0= No): {empleado[9]} Vacuna (1=Sí, 0= No): {empleado[10]}  Departamento: {empleado[11]}\n")
  
  id_cambio = str(input("Ingresa el id del empleado cuya información quieras modificar: "))

  for empleado in empleados:
    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    if id_cambio == id:
      existe = "si"

    else:
      existe = "no" 
  
  if existe == "no":
    print("El empleado no existe")

  else:
    print(f'Empleado encontrado, ID:    {empleado[0]}')
    for x in range(len(empleado)):
      id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado
      if empleado[x] == id:
        empleado[x] = int(input("Teclea el id del empleado: "))
      if empleado[x] == nombre:
        empleado[x] = str(input("Teclea el nombre del empleado: "))
      if empleado[x] == edad:
        empleado[x] = int(input("Teclea edad del empleado: "))
      if empleado[x] == peso:
        empleado[x] = int(input("Teclea el peso (Kg): "))
      if empleado[x] == estatura:
        empleado[x] = float(input("Teclea la estatura (Mts.): "))
      if empleado[x] == diabetes:
        empleado[x] = int(input("Diabetes 1-Sí 0-No ? : "))
      if empleado[x] == hipertension:
        empleado[x] = int(input("Hipertensión 1-Sí 0-No ? : "))
      if empleado[x] == corazon:
        empleado[x] = int(input("Enfermedad corazón 1-Sí 0-No ? : "))
      if empleado[x] == cancer:
        empleado[x] = int(input("Cáncer 1-Sí 0-No ? : "))
      if empleado[x] == tabaquismo:
        empleado[x] = int(input("Tabaquismo 1-Sí 0-No ? : ")
      if empleado[x] == vacuna: 
        empleado[x] = int(input("Vacuna Covid-19 1-Sí 0-No ? : "))
      if empleado[x] == departamento:
        empleado[x] = int(input("Teclea el número de departamento: ")) 
    print(empleado)   
    

def consultar_empleado():
  global empleados 
  
  print("Bienvenido a la opción para consultar un empleado, por favor sigue las instrucciones \nA continuación se despliega la lista de preguntas guardadas\n")

  for empleado in empleados:

    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    print(f"ID del empleado: {empleado[0]}  Nombre del empleado: {empleado[1]}  Edad del empleado: {empleado[2]}  Peso del empleado: {empleado[3]}  Estatura del empleado: {empleado[4]}  Diabetes (1 = Sí, 0= No): {empleado[5]} Hipertensión (1= Sí, 0= No): {empleado[6]}  Problemas del corazón (1=Sí, 0= No): {empleado[7]}  Cancer (1=Sí, 0= No): {empleado[8]} Tabaquismo (1=Sí, 0= No): {empleado[9]} Vacuna (1=Sí, 0= No): {empleado[10]}  Departamento: {empleado[11]}\n")
  
  id_cambio = str(input("Ingresa el id del empleado cuya información quieras modificar: "))

  for empleado in empleados:
    id, nombre, edad, peso, estatura, diabetes, hipertension, corazon, cancer, tabaquismo, vacuna, departamento = empleado

    if id_cambio == id:
      existe = "si"

    else:
      existe = "no" 
  
  if existe == "no":
    print("El empleado no existe")

  else:
    print(f'Empleado encontrado, {empleado}')

def crear_reporte_empleados():
  global empleados



def salir():
  print("Gracias por usar el sistema")
  si = False

def menu():

  print("Sistema de Salud y Bienestar del Empleado")
  print("1. Alta de empleado")
  print("2. Calcular imc de empleados")
  print("3. Cambia información del empleaddo")
  print("4. Consulta un empleado")
  print("5. Reporte de todos los empleados")
  print("6. Salir")
  opcion = int(input("Teclea la opcion: "))
  return opcion

def main():

  si = True 
  #matriz = []
  
  while si:

    opcion = menu()

    if opcion < 1 and opcion > 7:
      print("Opción incorrecta")
      menu()

    if opcion == 1:
      alta_empleado()

    elif opcion == 2:
      calcular_imc()
   
    elif opcion == 3: 
      actualizar_empleado()

    elif opcion == 4:
      consultar_empleado()
      
    elif opcion == 5:
      crear_reporte_empleados()

    elif opcion == 6:
      si = salir()
  
       
main()
```
